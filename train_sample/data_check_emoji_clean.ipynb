## See if contain emojis

```python
def contains_emoji(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.search(text) is not None
```

```python
rows_with_emojis = df[df.applymap(contains_emoji).any(axis=1)]
```

## Remove emoji and other special token

```python
def remove_emojis(text):
    # Regular expression to match all Emoji characters
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F700-\U0001F77F"  # alchemical symbols
        u"\U0001F780-\U0001F7FF"  # geometric shapes extended
        u"\U0001F800-\U0001F8FF"  # supplement arrows-C
        u"\U0001F900-\U0001F9FF"  # supplemental symbols and pictographs
        u"\U0001FA00-\U0001FA6F"  # chess symbols
        u"\U0001FA70-\U0001FAFF"  # symbols and pictographs extended-A
        u"\U00002700-\U000027BF"  # dingbats
        u"\U0001F7E0-\U0001F7EB"  # circle colors
        u"\U0001F90D-\U0001F93A"  # additional emoticons
        u"\U0001F93C-\U0001F945"  # sports & activities
        u"\U00002600-\U000026FF"  # miscellaneous symbols
        "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
df['train'] = df['train'].apply(remove_emojis)
```
