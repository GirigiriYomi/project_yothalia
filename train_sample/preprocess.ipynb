```python
import pandas as pd
from datasets import load_dataset

import re
```

```python
dataset = load_dataset("iamketan25/roleplay-instructions-dataset")
train_df = pd.DataFrame(dataset['train'])
test_df  =pd.DataFrame(dataset['test'])
```

```python
df = pd.concat([train_df,test_df])
```

```python
df.head()
```

```python
def split_instruction(prompt):
    prompt = prompt.replace('Human:','')
    return prompt

df['prompt'] = df['prompt'].map(split_instruction)
```

```python
df[['instruction', 'user']] = df['prompt'].str.split('.',n=1, expand=True)
df.drop(['prompt','__index_level_0__'],axis=1,inplace=True)
df.rename(columns={'chosen':'assisstant'},inplace=True)
```

```python
df = df[['instruction','user','assisstant']]
```

```python
df.head()
```

```python
df.to_csv('./csv/eng_roleplay.csv')
```

```python
instruction = []
user = []
assisstant = []
```

```python
with open('diverse_machine.txt','r',encoding='UTF-8') as file:
    while True:
        line = file.readline()
        if not line:
            break
        if line == '\n':
            continue
            
        line = line.strip().split('：',1)
        if line[0] == '指令':
            instruction.append(line[1])
        elif line[0] == '输入':
            user.append(line[1])
        elif line[0] == '输出':
            assisstant.append(line[1])
```

```python
with open('yothalia_machine.txt','r',encoding='UTF-8') as file:
    while True:
        line = file.readline()
        if not line:
            break
        if line == '\n':
            continue
            
        line = line.strip().split('：',1)
        print(line)
        if line[0] == '指令':
            instruction.append(line[1])
        elif line[0] == '输入':
            user.append(line[1])
        elif line[0] == '输出':
            assisstant.append(line[1])
```

```python
df_machine = pd.DataFrame({'instruction':instruction, 'user':user, 'assisstant':assisstant})
```

### Preprocess Tarklanse/Traditional_Chinese_roleplay_chat_Dataset

```python
from zhconv import convert
```

```python
dataset = load_dataset("Tarklanse/Traditional_Chinese_roleplay_chat_Dataset")
```

```python
train_df = pd.DataFrame(dataset['train'])
```

```python
train_df['instruction'] = train_df['instruction'].map(lambda x: x.replace('男性','').replace('雄性',''))
```

```python
empty_df = train_df.loc[train_df['input']=='']
train_df = train_df.loc[train_df['input']!='']
```

```python
split_pattern = r"以下.*?:"
empty_df[['instruction', 'input']] = empty_df['instruction'].str.split(split_pattern, expand=True)
```

```python
empty_df = empty_df.dropna()
```

```python
train_df = pd.concat([train_df,empty_df])
```

```python
train_df
```

```python
train_df = train_df.map(lambda x: convert(x,'zh-cn').strip())    
```

```python
train_df
```

```python
train_df.to_csv('./csv/zh_roleplay.csv')
```
